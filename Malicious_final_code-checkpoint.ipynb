```python
from sklearn.utils import resample
# Separate majority and minority classes
df_majority = data[data['type']== 0]
df_minority = data[data['type']== 1]
 
# Downsample majority class and upsample the minority class
df_minority_upsampled = resample(df_minority, replace=True,n_samples=200,random_state=1) 
df_majority_downsampled = resample(df_majority, replace=True,n_samples=200,random_state=1) 
 
# Combine minority class with downsampled majority class
df_balanced = pd.concat([df_minority_upsampled, df_majority_downsampled])
 
# Display new class counts
df_balanced['type'].value_counts()
```

```python
# shuffle the DataFrame rows 
df = df_balanced.sample(frac = 1)
df
```

## CNN

```python
# reshape the train data and test data
X_train = x_train.values.reshape(280, 330,1)
X_test = x_test.values.reshape(120, 330,1)
```

```python
X_train.shape
```

```python
from keras.layers import Dense, Dropout, Flatten, Conv1D, MaxPool1D, GlobalMaxPooling1D
```

```python
# Set the CNN model 

model = Sequential()

model.add(Conv1D(filters = 32, kernel_size = (5),padding = 'Same', activation ='relu', input_shape = ( 330, 1)))


model.add(Conv1D(filters = 32, kernel_size = (5),padding = 'Same', activation ='relu'))
model.add(MaxPool1D(pool_size=(2)))
model.add(Dropout(0.2))


model.add(Conv1D(filters = 64, kernel_size = (3),padding = 'Same', activation ='relu'))
model.add(Conv1D(filters = 64, kernel_size = (3),padding = 'same', activation ='relu'))
model.add(MaxPool1D(pool_size=(2), strides=(2)))
model.add(Dropout(0.3))

model.add(Conv1D(filters = 128, kernel_size = (3),padding = 'Same', activation ='relu'))
model.add(Conv1D(filters = 128, kernel_size = (3),padding = 'Same', activation ='relu'))
model.add(MaxPool1D(pool_size=(2), strides=(2)))
model.add(Dropout(0.4))



model.add(GlobalMaxPooling1D())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(2, activation = "softmax"))
model.summary()
```

```python
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
h1 = model.fit(X_train, y_train, epochs = 10, validation_data = (X_test,y_test),batch_size =2500)
```

```python
from sklearn.metrics import confusion_matrix

df_heatmap = pd.DataFrame(confusion_matrix(np.argmax((model.predict(X_test)),axis = 1),np.argmax(y_test,axis=1)))
heatmap = sns.heatmap(df_heatmap, annot=True, fmt="d")
```

```python
x1 = df.iloc[:,:-1]
y1 = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1, y1, test_size=0.3,stratify=y ,random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import GridSearchCV

dept = [1, 5, 10, 50, 100, 500, 1000]
n_estimators =  [20, 40, 60, 80, 100, 120]


param_grid={'n_estimators':n_estimators , 'max_depth':dept}
clf = RandomForestClassifier()
model = GridSearchCV(clf,param_grid,scoring='accuracy',n_jobs=-1,cv=3)
model.fit(x_train1,y_train1)
print("optimal n_estimators",model.best_estimator_.n_estimators)
print("optimal max_depth",model.best_estimator_.max_depth)
optimal_n_estimators = model.best_estimator_.n_estimators
optimal_max_depth = model.best_estimator_.max_depth
```

```python
#Testing AUC on Test data
rf = RandomForestClassifier(max_depth = optimal_max_depth,n_estimators =optimal_n_estimators)

rf.fit(x_train1,y_train1)

#predict on test data and train data
 
y_predtest = rf.predict(x_test1)
y_predtrain = rf.predict(x_train1)

print('*'*35)

#accuracy on training and testing data

print('the accuracy on testing data',accuracy_score(y_test1,y_predtest))
print('the accuracy on training data',accuracy_score(y_train1,y_predtrain))
train = accuracy_score(y_train1,y_predtrain)
test = accuracy_score(y_test1,y_predtest)

print('*'*35)


# Code for drawing seaborn heatmaps

cm = pd.DataFrame(confusion_matrix(y_test1, y_predtest.round()))
fig = plt.figure( )
heatmap = sns.heatmap(cm, annot=True, fmt="d")
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
# plt.figure(figsize(10,5))
error1= []
error2= []
for k in range(8,100):
    knn= KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train1,y_train1)
    y_pred1= knn.predict(x_train1)
    error1.append(np.mean(y_train1!= y_pred1))
    y_pred2= knn.predict(x_test1)
    error2.append(np.mean(y_test1!= y_pred2))
# plt.figure(figsize(10,5))
plt.plot(range(8,100),error1,label="train")
plt.plot(range(8,100),error2,label="test")
plt.xlabel('k Value')
plt.ylabel('Error')
plt.legend()
print(error1)
print(error2)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
import math

k = list(range(1,50,4))

train_acc = []
test_acc = []

for i in k:
    clf = KNeighborsClassifier(n_neighbors = i,algorithm='brute')
    clf.fit(x_train1,y_train1)
    pred_test = clf.predict(x_test1)
    test_acc.append(accuracy_score(y_test1,pred_test))
    pred_train = clf.predict(x_train1)
    train_acc.append(accuracy_score(y_train1,pred_train))
optimal_k = k[test_acc.index(max(test_acc))]
k = [math.log(x) for x in k]

#plot auc vs alpha
x = plt.subplot( )
x.plot(k, train_acc, label='Train Accuracy')
x.plot(k, test_acc, label='Test Accuracy')
plt.title('Accuracy vs hyperparameter')
plt.xlabel('k')
plt.ylabel('Accuracy')
x.legend()
plt.show()

print('optimal alpha for which auc is maximum : ',optimal_k)
```

```python
#Testing AUC on Test data
knn = KNeighborsClassifier(n_neighbors = optimal_k,algorithm='brute')
knn.fit(x_train1,y_train1)

#predict on test data and train data
 
y_predtestk = knn.predict(x_test1)
y_predtraink = knn.predict(x_train1)

print('*'*35)

#accuracy on training and testing data

print('the accuracy on testing data',accuracy_score(y_test1,y_predtestk))
print('the accuracy on training data',accuracy_score(y_train1,y_predtraink))
train1 = accuracy_score(y_train1,y_predtraink)
test1 = accuracy_score(y_test1,y_predtestk)

print('*'*35)


# Code for drawing seaborn heatmaps

cm = pd.DataFrame(confusion_matrix(y_test1, y_predtestk.round()))
fig = plt.figure( )
heatmap = sns.heatmap(cm, annot=True, fmt="d")
```

```python
from sklearn.svm import SVC
from sklearn.calibration import CalibratedClassifierCV
import math
C = [10000,1000,100,10,1,0.1,0.01,0.001,0.0001]

train_auc = []
cv_auc = []

for i in C:
    model = SVC(C=i,gamma=50) 
    clf = CalibratedClassifierCV(model, cv=3)
    clf.fit(x_train1,y_train1)
    prob_cv = clf.predict(x_test1)
    cv_auc.append(accuracy_score(y_test1,prob_cv))
    prob_train = clf.predict(x_train1)
    train_auc.append(accuracy_score(y_train1,prob_train))
optimal_C= C[cv_auc.index(max(cv_auc))]
C=[math.log(x) for x in C]

#plot auc vs alpha
x = plt.subplot( )
x.plot(C, train_auc, label='AUC train')
x.plot(C, cv_auc, label='AUC CV')
plt.title('AUC vs hyperparameter')
plt.xlabel('C')
plt.ylabel('AUC')
x.legend()
plt.show()

print('optimal C for which auc is maximum : ',optimal_C)
```

```python
gamma = [10,20,30,30,40]

train_auc = []
cv_auc = []

for i in gamma:
    model = SVC(C=1000,gamma=i) 
    clf = CalibratedClassifierCV(model, cv=3)
    clf.fit(x_train1,y_train1)
    prob_cv = clf.predict(x_test1)
    cv_auc.append(accuracy_score(y_test1,prob_cv))
    prob_train = clf.predict(x_train1)
    train_auc.append(accuracy_score(y_train1,prob_train))
optimal_gamma= gamma[cv_auc.index(max(cv_auc))]
# C=[math.log(x) for x in C]

#plot auc vs alpha
x = plt.subplot( )
x.plot(gamma, train_auc, label='AUC train')
x.plot(gamma, cv_auc, label='AUC CV')
plt.title('AUC vs hyperparameter')
plt.xlabel('gamma')
plt.ylabel('AUC')
x.legend()
plt.show()

print('optimal gamma for which auc is maximum : ',optimal_gamma)
```

```python
#Testing AUC on Test data
svc = SVC(C=optimal_C,gamma=optimal_gamma)

svc.fit(x_train1,y_train1)

#predict on test data and train data
 
y_predtests = svc.predict(x_test1)
y_predtrains = svc.predict(x_train1)

print('*'*35)

#accuracy on training and testing data

print('the accuracy on testing data',accuracy_score(y_test1,y_predtests))
print('the accuracy on training data',accuracy_score(y_train1,y_predtrains))
train2 = accuracy_score(y_train1,y_predtrains)
test2 = accuracy_score(y_test1,y_predtests)

print('*'*35)


# Code for drawing seaborn heatmaps

cm = pd.DataFrame(confusion_matrix(y_test1, y_predtests.round()))
fig = plt.figure( )
heatmap = sns.heatmap(cm, annot=True, fmt="d")
```
